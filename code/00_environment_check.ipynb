## 1. 실행 환경 확인

파이썬 및 핵심 라이브러리의 설치 상태를 확인합니다. 오류 없이 버전이 출력되면 정상입니다.

```python
import sys
print('Python', sys.version.split()[0])

import numpy as np
import matplotlib.pyplot as plt
print('numpy', np.__version__)
```

## 1-1. 그래프 한글 폰트 설정

그래프 제목·축 라벨에 한글을 쓸 때 깨지지 않도록, 설치된 한글 폰트를 찾아 적용합니다.

```python
import matplotlib.font_manager as fm
for name in ['Malgun Gothic','NanumGothic','AppleGothic','Noto Sans CJK KR','Noto Sans KR']:
    if name in {f.name for f in fm.fontManager.ttflist}:
        plt.rcParams['font.family'] = name; print('한글 폰트 적용:', name); break
else:
    print('한글 폰트를 찾지 못했습니다 — 라벨은 영어로 표기합니다.')
plt.rcParams['axes.unicode_minus'] = False
```

## 2. 예제 이미지 생성

실제 TEM 파일을 사용하기 전, 수치로 구성된 예제 이미지를 생성합니다.
이미지는 격자 형태로 배열된 수치 데이터이며, 상세 내용은 Phase 3 에서 다룹니다.

```python
# 200 x 300 크기의 예제 이미지 (밝기값 0~255)
img = np.zeros((200, 300), dtype=np.uint8)
img[60:140, :] = 180      # 중앙 가로 영역을 밝게 설정
img[90:110, :] = 90       # 내부에 어두운 층 설정

print('shape :', img.shape)   # (행, 열) = (높이, 너비)
print('dtype :', img.dtype)
```

## 3. 이미지 표시

`matplotlib` 으로 이미지를 표시합니다. 가로 방향의 밝은 영역과 내부의 어두운 층이 표시되면 정상입니다.

```python
plt.figure(figsize=(5,3))
plt.imshow(img, cmap='gray')
plt.title('sample image')
plt.axis('off')
plt.show()
```
