```python
import numpy as np
```

### 문제 1. 배열 만들기와 shape
`arange(12)` 로 0~11 을 만들고 `reshape(3,4)` 로 3행 4열로 바꿉니다.

```python
arr = np.arange(12).reshape(3, 4)

assert arr.shape == (3, 4)
assert arr[0, 0] == 0 and arr[2, 3] == 11
print('통과')
```

### 문제 2. ROI 슬라이싱
가운데 2x2 는 행 2~3, 열 2~3 이므로 `[2:4, 2:4]` 입니다.

```python
g = np.arange(36).reshape(6, 6)
roi = g[2:4, 2:4]

assert roi.shape == (2, 2)
assert roi.tolist() == [[14, 15], [20, 21]]
print('통과')
```

### 문제 3. ROI 값 바꾸기
행 0~1 전체는 `[0:2, :]` 로 선택해 0을 대입합니다.

```python
g2 = np.arange(36).reshape(6, 6)
g2[0:2, :] = 0

assert g2[0].sum() == 0 and g2[1].sum() == 0
assert g2[2].sum() != 0
print('통과')
```

### 문제 4. 불리언 마스킹 — 개수 세기
`img > 100` 마스크의 True 개수를 셉니다.

```python
img = np.array([[ 10, 160,  80],
                [200,  50, 255],
                [120, 240,  30]], dtype=np.uint8)
cnt = np.count_nonzero(img > 100)

assert cnt == 5
print('통과')
```

### 문제 5. 마스킹으로 값 바꾸기
`out[out <= 100] = 0` 으로 조건 픽셀만 0으로 바꿉니다.

```python
out = img.copy()
out[out <= 100] = 0

assert out.tolist() == [[0, 160, 0], [200, 0, 255], [120, 240, 0]]
print('통과')
```

### 문제 6. axis 평균
열별 평균은 `axis=0` 입니다.

```python
a = np.array([[2, 4, 6],
              [4, 6, 8]], dtype=float)
col_mean = a.mean(axis=0)

assert col_mean.tolist() == [3.0, 5.0, 7.0]
print('통과')
```

### 문제 7. 값이 아니라 위치 찾기 (argmax)
`argmax` 는 최댓값의 위치(인덱스)를 반환합니다. 동일 최댓값이 여러 개면 첫 번째 위치.

```python
v = np.array([5, 30, 8, 30, 12])
pos = v.argmax()

assert pos == 1
print('통과')
```
