## 1. 검증용 예제 이미지 (알려진 정답)
AI 코드를 검증하려면 **정답을 아는 시료**가 필요합니다. 두께 35px 인 밝은 층 이미지를 만듭니다. scale=0.25 이므로 정답 두께는 35 × 0.25 = 8.75 nm 입니다.

```python
import numpy as np, cv2
np.random.seed(3)
H,W=220,360; img=np.full((H,W),90,float); img[95:130,:]=185
img=np.clip(img+np.random.normal(0,8,(H,W)),0,255).astype(np.uint8)
scale=0.25
print('참 두께: 35 px =', 35*scale, 'nm')
```

### 문제 1. AI 코드 검증하기
함수를 호출해 반환값을 확인합니다. 밝은 층 범위 안이면 신뢰할 수 있습니다.

```python
def brightest_row_ai(img):
    return int(img.mean(axis=1).argmax())

row = brightest_row_ai(img)

assert 95 <= row < 130
print('통과 — 가장 밝은 행:', row)
```

### 문제 2. 버그 찾아 고치기 (scale 누락)
px 결과에 scale 을 곱해 nm 로 환산합니다.

```python
def measure_px(img, x):
    prof=img[:,x].astype(float); level=(prof.min()+prof.max())/2
    cr=[]
    for i in range(len(prof)-1):
        if (prof[i]-level)*(prof[i+1]-level)<0:
            t=(level-prof[i])/(prof[i+1]-prof[i]); cr.append(i+t)
    return cr[-1]-cr[0]

def measure_nm(img, x, scale):
    return measure_px(img, x) * scale

assert abs(measure_nm(img, 180, 0.25) - 8.75) < 0.3
print('통과')
```

### 문제 3. 검증 함수 작성
절댓값 오차가 허용치 이하인지 판정합니다.

```python
def verify(value, expected, tol):
    return abs(value - expected) <= tol

assert verify(8.7, 8.75, 0.3) is True
assert verify(6.0, 8.75, 0.3) is False
print('통과')
```
