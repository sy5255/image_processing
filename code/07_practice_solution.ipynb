## 1. 데이터 준비 (먼저 실행)

```python
import pandas as pd, numpy as np
np.random.seed(11)
n=30
thick=np.round(np.random.normal(8.75,0.3,n),2); thick[5]=10.9; thick[20]=6.7
df=pd.DataFrame({'pos':np.arange(1,n+1),'thick_nm':thick,'sample':['A']*15+['B']*15})
print('df 준비 완료:', df.shape)
```

### 문제 1. 평균·표준편차
Series 의 mean(), std() 를 사용합니다.

```python
m = df['thick_nm'].mean()
sd = df['thick_nm'].std()

assert abs(m - df['thick_nm'].mean()) < 1e-9
assert abs(sd - df['thick_nm'].std()) < 1e-9
print('통과')
```

### 문제 2. 행 필터
불리언 조건으로 행을 고릅니다.

```python
big = df[df['thick_nm'] >= 9.0]

assert (big['thick_nm'] >= 9.0).all()
assert len(big) == (df['thick_nm']>=9.0).sum()
print('통과')
```

### 문제 3. groupby 평균
groupby 로 시료별 평균을 구합니다.

```python
g = df.groupby('sample')['thick_nm'].mean()

assert abs(g['A'] - df[df['sample']=='A']['thick_nm'].mean()) < 1e-9
print('통과')
```

### 문제 4. IQR 이상치 제거
IQR 범위(lo~hi) 안의 행만 남깁니다.

```python
q1=df['thick_nm'].quantile(0.25); q3=df['thick_nm'].quantile(0.75); iqr=q3-q1
lo,hi = q1-1.5*iqr, q3+1.5*iqr
clean = df[(df['thick_nm']>=lo) & (df['thick_nm']<=hi)]

assert clean['thick_nm'].max() <= hi and clean['thick_nm'].min() >= lo
assert len(clean) < len(df)
print('통과 — 남은 행', len(clean))
```

### 문제 5. CSV 저장
to_csv 로 저장하며 index=False 로 인덱스를 제외합니다.

```python
import tempfile, os
path=os.path.join(tempfile.gettempdir(),'out.csv')
clean.to_csv(path, index=False)

assert os.path.exists(path)
print('통과')
```
