# Phase 1 — 파이썬 기초 문법 (예제)

계측 코드 작성에 필요한 핵심 문법을 정리한 예제입니다. 각 셀을 직접 실행하고, 값을 변경하며 결과를 확인하십시오.

## 1. 변수와 자료형
변수는 값을 저장하는 공간이며, `=` 기호로 값을 할당합니다. 자료형은 정수(int) · 실수(float) · 문자열(str) · 논리값(bool) 4종을 사용합니다.

```python
scale = 0.25          # float : nm/pixel
count = 12            # int   : 개수
name  = 'layer_A'     # str   : 문자열
ok    = True          # bool  : 논리값

print(scale, count, name, ok)
print(type(scale), type(count), type(name), type(ok))
```

## 2. list · tuple · dict
여러 값을 하나의 변수에 저장합니다.
- **list** `[ ]` 순서 있음, 수정 가능 — 측정값 모음
- **tuple** `( )` 수정 불가 — 고정 좌표 (x, y)
- **dict** `{ }` 이름:값 쌍 — 단일 측정의 속성 집합

```python
widths = [12.1, 11.8, 12.4]           # list
point  = (128, 64)                   # tuple : (x, y)
result = {'name': 'A', 'thk': 12.4}  # dict

print(widths)
print(point)
print(result['thk'])     # dict 는 이름으로 값을 참조
```

## 3. 자료구조의 값 추가 · 수정 · 제거
측정 결과를 누적하고 정리하기 위한 핵심 연산입니다.
- **list**: `append` 추가, 인덱스 대입으로 수정, `remove`/`pop` 제거
- **dict**: `d[키] = 값` 으로 추가·수정, `del` 로 제거
- **tuple**: 불변(immutable). 추가·수정·제거 불가하며, 변경이 필요하면 새 튜플을 생성

```python
# list : 추가 · 수정 · 제거
w = [12.1, 11.8]
w.append(12.4)      # 끝에 추가      -> [12.1, 11.8, 12.4]
w[0] = 12.0         # 인덱스로 수정  -> [12.0, 11.8, 12.4]
w.remove(11.8)      # 값으로 제거    -> [12.0, 12.4]
removed = w.pop(0)  # 위치로 제거하며 값 반환
print(w, '| removed =', removed)
```

```python
# dict : 추가 · 수정 · 제거
d = {'name': 'A'}
d['thk'] = 12.4     # 키 추가
d['thk'] = 13.0     # 동일 키 재할당 = 수정
del d['name']       # 키 제거
print(d)            # {'thk': 13.0}
```

```python
# tuple : 불변 — 변경하려면 새로 생성
p = (128, 64)
# p[0] = 130        # 오류 발생 (수정 불가)
p = (130, 64)       # 새 튜플로 교체
print(p)
```

## 4. 인덱싱 및 슬라이싱
자료구조에서 원하는 부분을 선택합니다. 인덱스는 0부터 시작합니다. 이 문법은 이미지 ROI 선택의 기초가 됩니다.

```python
w = [12.1, 11.8, 12.4, 13.0]
print(w[0])     # 12.1  첫 번째
print(w[-1])    # 13.0  마지막
print(w[1:3])   # [11.8, 12.4]  1번 이상 3번 미만
print(w[:2])    # [12.1, 11.8]  처음 2개
```

## 5. 조건문 (if · elif · else)
조건에 따라 분기하여 실행합니다. 들여쓰기(공백 4칸)로 코드 블록을 구분하는 것이 파이썬의 기본 규칙입니다.

```python
thk = 12.4

if thk > 13:
    print('두꺼움')
elif thk < 11:
    print('얇음')
else:
    print('정상')
```

## 6. 반복문 (for · range)
동일한 처리를 반복 수행합니다. `for` 는 자료구조의 요소를 순회하고, `range` 는 지정 횟수만큼 반복합니다.

```python
widths = [12.1, 11.8, 12.4]
for w in widths:
    print(w * 0.25)     # 픽셀 -> nm 환산

print('---')
for i in range(3):
    print(i)            # 0, 1, 2
```

## 7. 반복문 (while)
조건이 참인 동안 반복 수행합니다. 조건 갱신 구문을 누락하면 반복이 종료되지 않으므로 유의하십시오.

```python
i = 0
while i < 3:
    print(i)
    i = i + 1      # 이 구문이 없으면 종료되지 않음
```

## 8. enumerate · zip · continue · break
반복을 보다 효율적으로 처리하는 도구입니다.
- **enumerate** 인덱스와 값 동시 제공
- **zip** 두 자료구조를 짝지어 순회
- **continue** 현재 반복 건너뜀
- **break** 반복 즉시 종료

```python
widths = [12.1, None, 12.4, 13.0]   # None = 측정 실패

for i, w in enumerate(widths):
    if w is None:
        continue        # 실패값은 건너뜀
    print(i, w)

print('---')
xs = [10, 20, 30]
ys = [64, 66, 65]
for x, y in zip(xs, ys):
    print(x, y)         # 좌표 짝짓기
```

## 9. 함수 (def · return)
코드에 이름을 부여하여 재사용합니다. `def` 로 정의하고, 인자를 입력받아 `return` 으로 결과를 반환합니다.

```python
def px_to_nm(px, scale):
    return px * scale

print(px_to_nm(50, 0.25))     # 12.5
print(px_to_nm(128, 0.25))    # 32.0
```

## 10. 적용 예: 측정값 평균 계산
측정값 리스트의 평균을 계산하는 함수입니다. list · for · 함수가 함께 사용됩니다.
본 단계에서는 평균 계산을 직접 구현하며, 이후 NumPy 의 `np.mean` 으로 대체합니다.

```python
def average(values):
    total = 0
    for v in values:
        total = total + v
    return total / len(values)

widths = [12.1, 11.8, 12.4, 13.0]
print(average(widths))     # 12.325
```
